## Desafío - Tipos y Estructuras de Datos (Parte II)

```python
# Importando pandas

import pandas as pd

# 1) Creando dataframe con Archivo astronautas csv

df = pd.read_csv("resumen_resultados_astronautas.csv")
df.head()

# 2) Creando columna IMC

# Cálculo correcto del IMC
df["IMC"] = df["peso"] / ((df["altura"] / 100) ** 2)
df.head()

# 3) Creando columna Promedio Evaluaciones

eval_cols = ["evaluacion_fisica", "evaluacion_psicologica", "evaluacion_adaptativa", "evaluacion_medica"]
df["promedio_evaluaciones"] = df[eval_cols].mean(axis=1)
df.head()

# 4) Filtrando IMC dentro de limites entregados

df_filtrado = df[(df["IMC"] >= 18.5) & (df["IMC"] <= 24.9)] 

# 5) Filtrando por promedio de evaluaciones >87

df_filtrado = df_filtrado[df_filtrado["promedio_evaluaciones"] >= 87]

# 6) Reiniciando el indice

df_filtrado = df_filtrado.reset_index(drop=True)


# 7) Modificando el valor de la columna "califica", para que diga "Si"

df_filtrado["califica"] = "Sí"


# 8) Exportando resultados a csv
df_filtrado.to_csv("astronautas_calificados.csv", index=False)

```

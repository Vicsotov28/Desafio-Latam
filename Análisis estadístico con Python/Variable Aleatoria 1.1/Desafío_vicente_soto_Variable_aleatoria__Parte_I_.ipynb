```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar dataset
df = pd.read_csv("world-data-2023.csv")

# Inspección inicial
df.head()
```

## 2. Información general del dataset

Revisamos información sobre las columnas, los tipos de datos y el conteo de valores nulos.


```python
df.info()
```

## 3. Estadísticas iniciales

Obtenemos estadísticas descriptivas de las variables numéricas y categóricas.


```python
df.describe(include="all")
```

## 4. Limpieza de columnas irrelevantes

Excluimos del análisis las siguientes columnas:  
- Abbreviation  
- Calling Code  
- Capital/Major City  
- Currency_Code  
- Largest City  
- Official Language  
- Latitude  
- Longitude  


```python
cols_drop = [
    "Abbreviation", 
    "Calling Code", 
    "Capital/Major City",
    "Currency-Code",     
    "Largest city",      
    "Official language", 
    "Latitude", 
    "Longitude"
]

df = df.drop(columns=cols_drop)
df.head()
```

## 5. Análisis exploratorio inicial

Generamos histogramas para observar la distribución de las variables numéricas.


```python
df.hist(figsize=(15,10), bins=30)
plt.tight_layout()
plt.show()
```

## 6. Correlaciones entre variables

Construimos un mapa de calor para visualizar las correlaciones entre variables numéricas.


```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
plt.title("Mapa de calor de correlaciones")
plt.show()
```

## 7. Selección de pares de variables y limpieza

A partir del mapa de calor, seleccionamos dos pares de variables con correlación alta:  
- **Población vs PIB**  
- **Médicos por mil habitantes vs Esperanza de vida**  

Aplicamos limpieza (eliminación de nulos y outliers) para mejorar la calidad del análisis.


```python
# --- Limpieza de datos ---

df["GDP"] = (
    df["GDP"]
    .replace(r"[\$,]", "", regex=True)  # quitar $ y comas
    .astype(float)
)

df["Population"] = (
    df["Population"]
    .replace(r"[\$,]", "", regex=True)  # quitar comas
    .astype(float)
)

# --- Subconjunto GDP y Population ---
subset_gdp_pop = df[["Population", "GDP"]].dropna()

# --- Eliminación de outliers con IQR ---
Q1 = subset_gdp_pop["GDP"].quantile(0.25)
Q3 = subset_gdp_pop["GDP"].quantile(0.75)
IQR = Q3 - Q1

subset_gdp_pop = subset_gdp_pop[
    (subset_gdp_pop["GDP"] >= Q1 - 1.5*IQR) & 
    (subset_gdp_pop["GDP"] <= Q3 + 1.5*IQR)
]

subset_gdp_pop.head()
```

## 8. Visualización de relaciones

Graficamos los pares seleccionados para observar las relaciones entre variables.


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=subset_gdp_pop, x="Population", y="GDP")
plt.title("Relación entre Población y PIB (limpios)")
plt.show()
```

```python
subset_med_life = df[["Physicians per thousand", "Life expectancy"]].dropna()

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=subset_med_life,
    x="Physicians per thousand",
    y="Life expectancy"
)
plt.title("Relación entre médicos por cada mil habitantes y expectativa de vida")
plt.xlabel("Médicos por cada mil habitantes")
plt.ylabel("Expectativa de vida (años)")
plt.show()
```

## 9. Caso especial: Escalas logarítmicas

Para datos muy dispersos como Población y PIB, se recomienda usar escala log-log.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Population", y="GDP")
plt.xscale("log")
plt.yscale("log")
plt.title("Relación entre Población y PIB (escala log)")
plt.show()
```

## 10. Análisis libre

Exploramos otra relación interesante: **Precio de la gasolina vs PIB**.

```python
subset_gas_gdp = df[["Gasoline Price", "GDP"]].dropna()

plt.figure(figsize=(8,6))
sns.scatterplot(data=subset_gas_gdp, x="Gasoline Price", y="GDP")
plt.title("Precio de la gasolina vs PIB")
plt.show()
```

# ✅ Conclusiones
- La limpieza de columnas irrelevantes simplificó el análisis.
- El mapa de calor permitió seleccionar variables con correlación significativa.
- El uso de escalas logarítmicas ayudó a visualizar datos con alta dispersión.
- Existen relaciones claras, por ejemplo, entre mayor cantidad de médicos y mayor esperanza de vida.
